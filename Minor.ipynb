```python
import psycopg2 as pg_driver

db = pg_driver.connect(
    database="postgres", 
    user='postgres',
    password='postgres', 
    host='localhost', 
    port='5432'
)

# Далее нам нужен курсор, который позволит нам непосредственно отправлять запросы на выполнения в БД. Эта строка всегда одинаковая
cur = db.cursor()
# Делаем запрос к таблице по умолчанию pg_database
cur.execute("select * from pg_database")

# Чтобы извлечь результат выполнения запроса к таблице по умолчанию, необходимо воспользоваться методом .fetchall()
rs = cur.fetchall()


# enumerate позволяет красиво пронумеровать строки
for row, values in enumerate(rs):
    print(row, values)
```

```python
import psycopg2 as pg_driver

db = pg_driver.connect(
                        database="flowers_db", 
                        user='pruser',
                        password='pruser', 
                        host='localhost', 
                        port='5432'
                    )



def execute_queries(db, sql_commands):
    db.autocommit = True
    with db.cursor() as cursor:
        for sql_command in sql_commands:
            print(sql_command)
            cursor.execute(sql_command)


sql_commands = ["DROP TABLE IF EXISTS seller;",
                "DROP TABLE IF EXISTS cars;",
                "DROP TABLE IF EXISTS cars_drivers;",
                "DROP TABLE IF EXISTS driver;",
                "DROP TABLE IF EXISTS type_of_flower;",
                "DROP TABLE IF EXISTS flower;",
                "DROP TABLE IF EXISTS bouquet;",
                "DROP TABLE IF EXISTS orders;",
                "DROP TABLE IF EXISTS customer;",
                "DROP TABLE IF EXISTS reviews;", 
                """CREATE TABLE seller (
                         id    INT       NOT NULL,
                         name    VARCHAR(64)       NOT NULL,
                         bank_details  VARCHAR(32)  NOT NULL
                         
                );
                """,
                """CREATE TABLE cars (
                         id    INT       NOT NULL,
                         car_num    VARCHAR(32)       NOT NULL,
                         car_color    VARCHAR(32)       NOT NULL,
                         car_brand    VARCHAR(32)       NOT NULL,
                         seller_id    INT       NOT NULL
                         
                  );
                """,
                """CREATE TABLE cars_drivers (
                         id    INT       NOT NULL,
                         car_id    INT       NOT NULL,
                         driver_id    INT       NOT NULL
                         
                );
                """,
                """CREATE TABLE driver (
                         id    INT       NOT NULL,
                         name    VARCHAR(64)       NOT NULL,
                         passport    VARCHAR(128)       NOT NULL
                         
                );
                """,
                """CREATE TABLE type_of_flower (
                         id    INT       NOT NULL,
                         name    VARCHAR(64)       NOT NULL,
                         seller_id    INT       NOT NULL,
                         date_of_delivery   DATE        NOT NULL,
                         cost_per_one    FLOAT       NOT NULL
                         
                );
                """,
                """CREATE TABLE flower (
                         id    INT       NOT NULL,
                         color    VARCHAR(32)       NOT NULL,
                         bouquet_id    INT       NOT NULL,
                         type_id    INT       NOT NULL
                         
                );
                """,
                """CREATE TABLE bouquet (
                         id    INT       NOT NULL,
                         name    VARCHAR       NOT NULL,
                         cost    FLOAT       NOT NULL,
                         occasion    VARCHAR(64)       NOT NULL
                         
                );
                """,
                """CREATE TABLE customer (
                         id    INT       NOT NULL,
                         name    VARCHAR(64)       NOT NULL,
                         phone    VARCHAR(11)       NOT NULL,
                         email    VARCHAR(128)       NOT NULL,
                         address    VARCHAR(256)       NOT NULL
                         
                );
                """,
                """CREATE TABLE orders (
                         id    INT       NOT NULL,
                         customer_id    INT       NOT NULL,
                         bouquet_id    INT       NOT NULL
                         
                );
                """,
                """CREATE TABLE reviews (
                         id    INT       NOT NULL,
                         order_id    INT       NOT NULL,
                         date   DATE        NOT NULL,
                         text   TEXT        NOT NULL       
                );
                """]

execute_queries(db, sql_commands)
```
